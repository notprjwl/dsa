```python
#find the length of the longest subsequence of two sequences
def lcs_recursive(seq1, seq2, idx1 = 0, idx2 = 0):
    if(idx1==len(seq1) or (idx2==len(seq2))):
        return 0
    elif seq1[idx1] == seq2[idx2]:
        return 1 + lcs_recursive(seq1, seq2, idx1+1, idx2+1)
    else:
        option1 = lcs_recursive(seq1, seq2, idx1+1, idx2)
        option2 = lcs_recursive(seq1, seq2, idx1, idx2+1)
        return max(option1,option2)


lcs_recursive('serendipitous', 'precipitation')
```

```python
#faster using memoization technique - it is used to speed up the programs
#for clear output(in ms) refer notebook from jovian
def lcs_memo(seq1, seq2):
    memo = {}
    def recursive(idx1, idx2):
        key = (idx1, idx2)
        if key in memo:
            return memo[key]
        elif idx1 == len(seq1) or idx2 == len(seq2):
            memo[key] = 0
        elif seq1[idx1] == seq2[idx2]:
            memo[key] = 1 + recursive(idx1+1, idx2+1)
        else:
            memo[key] = max(recursive(idx1+1, idx2), recursive(idx1, idx2+1))
        return memo[key]
    return recursive(0, 0)

lcs_memo('serendipitous', 'precipitation')
```

```python
def lcs(seq1, seq2):
    #creating the matrix of rocs as seq1 and columns as seq2
    dp = [[0 for j in range(len(seq2)+1)] for i in range(len(seq1)+1)]   

    for i in range(len(seq1) - 1, -1, -1):
        for j in range(len(seq2) - 1, -1, -1):
            if seq1[i]==seq2[j]:
                dp[i][j] = 1 + dp[i+1][j+1]
            else:
                dp[i][j] = max(dp[i][j+1], dp[i+1][j])
    
    return dp[0][0]
                
lcs('serendipitous', 'precipitation')
```

```python
#using recursion (naive recursive solution)

def Knapsack(n, c, w, value):         #n is the index and c is the capacity
    if n==0 or c==0:
        return 0
    elif w[n-1] > c:
        return Knapsack(n-1, c, w, value)
    else:
        return max(value[n-1] + Knapsack(n-1, c-w[n-1], w, value), Knapsack(n-1, c, w, value))

weight = [10, 20, 30]
profit = [60, 100, 120]
capacity = 50
n = len(profit)
print(Knapsack(n, capacity, weight, profit))        #120+100=220
```
